```python
a=[1,2,3,4,5]
sum=0
b=[]
for i in range(len(a)):
    sum=sum+a[i]
    b.append(sum)
print(b)    
```

```python
a=[1,0,1,0,1,0,0,0,1,1,0,1,1]
count=0
b=[]
for i in range(len(a)):
    if a[i]!=0:
        b.append(a[i])
    else:
        count+=1    


for j in range(count):
    b.append(0)


print(b)
```
