```python
l = [1,1,1,2,3,4,4,5,5,76,4,54,5,33,3,3]  # A list with duplicate values
setter = set(l)                          # Converts the list to a set (removes duplicates)
print(setter)                            # Prints the set
```

```python
l = [2,3,4,4,5,5,76,4,54,5,33,3,3]
sets = set(l)
sets.discard(54)
print(sets)
sets.discard(54)
print(sets.pop())
print(sets)
```
